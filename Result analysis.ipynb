```python
import numpy
import matplotlib.pyplot as plt
```

```python
labels = ['Normal', 'GN', 'SN', 'GB', 'Brightness', 'Hue', 'Contrast']
acc_resnet = [70.88, 61.09, 41.65, 62.21, 43.30, 48.11, 36.96]
acc_jointnet = [58.38, 49.95, 33.27, 52.24, 36.55, 39.61, 31.11]
acc_ttt = [58.21, 50.02, 32.95, 51.09, 37.34, 39.84, 39.48]
```

```python
# set width of the bars
bar_width = 0.25

# set positions of the bars on the x-axis
r1 = range(len(labels))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# plot the bars
plt.bar(r1, acc_resnet, color='coral', width=bar_width, edgecolor='white', label='ResNet')
plt.bar(r2, acc_jointnet, color='lightseagreen', width=bar_width, edgecolor='white', label='Joint net')
plt.bar(r3, acc_ttt, color='goldenrod', width=bar_width, edgecolor='white', label='TTT')

# add xticks on the middle of the group bars
plt.xlabel('Noise type')
plt.ylabel('Accuracy')
plt.xticks([r + bar_width for r in range(len(labels))], labels)
plt.title('Accuracy under different noisy data set')

# add legend
plt.legend()

# show the plot
plt.show()
plt.savefig('acc plot.png', dpi=800)
```

```python
change1 = [acc_resnet[0] - acc_resnet[i] for i in range(1, 7)]
change2 = [acc_jointnet[0] - acc_jointnet[i] for i in range(1, 7)]
change3 = [acc_ttt[0] - acc_ttt[i] for i in range(1, 7)]
```

```python
# set width of the bars
bar_width = 0.25

# set positions of the bars on the x-axis
r1 = range(len(labels[1:]))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# plot the bars
plt.bar(r1, change1, color='coral', width=bar_width, edgecolor='white', label='ResNet')
plt.bar(r2, change2, color='lightseagreen', width=bar_width, edgecolor='white', label='Joint net')
plt.bar(r3, change3, color='goldenrod', width=bar_width, edgecolor='white', label='TTT')

# add xticks on the middle of the group bars
plt.xlabel('Noise type')
plt.ylabel('Accuracy decrease')
plt.xticks([r + bar_width for r in range(len(labels[1:]))], labels[1:])
plt.title('Accuracy decrease under different noisy data set')

# add legend
plt.legend()

# show the plot
plt.show()
```
